```python
import pandas as pd

# dataset sera esta

dato1 = [1,2,2,3,4,4,5,6]
dato2 = [2,3,4,4,4,6,5,7]

diccionario = { "x": dato1, "y": dato2}

df = pd.DataFrame(diccionario)

df





```

```python
df ["xy"] = df.x * df.y
df

```

```python
df ["x2"] = df.x **2
df
```

```python
# Calcular la suma de cada columna
sumas_columnas = df[['x', 'y', 'xy', 'x2']].sum()

# Crear un DataFrame con la fila de totales
df_totales = pd.DataFrame([sumas_columnas], columns=df.columns)

# Asignar un índice significativo a la fila de totales (opcional, pero útil)
df_totales.index = ['Total']

# Concatenar el DataFrame original con la fila de totales
df_con_totales = pd.concat([df, df_totales])

# Mostrar el DataFrame resultante
display(df_con_totales)
```
